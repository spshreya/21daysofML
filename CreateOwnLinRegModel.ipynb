```
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
```

```
xs=np.array([1,2,3,4,5,6], dtype=np.float64)
ys=np.array([5,4,6,5,6,7],dtype=np.float64)
plt.scatter(xs,ys)
```

image.png

```
def bestFitSlope(xs,ys):
  m= (((mean(xs)*mean(ys))-mean(xs*ys))/ ((mean(xs)*mean(xs)) - mean(xs*xs)))



  return m
```

```
m = bestFitSlope(xs, ys)
m
```

image.png

```
def intercept(xs,ys,m):

  b= mean(ys)- m* mean(xs)
  return b
  
```

```
b = intercept(xs,ys,m)
b
```

```
regressionLine = [(m*x)+b for x in xs]
```

```
from matplotlib import style
style.use('fivethirtyeight')
plt.scatter(xs,ys)
plt.plot(xs,regressionLine)
```

```
predict_x = 8
predict_y = m*predict_x + b

plt.scatter(predict_x, predict_y, color = 'g')
plt.scatter(xs,ys)
plt.plot(xs,regressionLine)
```

```
def squaredError(ys_original, ys_line):



  return sum((ys_line-ys_original)**2)


def coeff(ys_original, ys_line):
  y_mean_line = [mean(ys_original) for y in ys_original]
  squared_error_reg= squaredError(ys_original, ys_line)
  squared_error_ymean =squaredError (ys_original, y_mean_line)

  return 1-(squared_error_reg/squared_error_ymean)
```

```
r_squared = coeff(ys, regressionLine)
r_squared  #if its zero it is equal to the y_mean line
```

```
import random
def create_dataset(hm, variance, step=2, correlation=False):
  val =1
  ys=[]
  for i in range(hm):
    y = val + random.randrange(-variance, variance)
    ys.append(y)
    if correlation == 'pos':
      val+=step
    elif correlation == 'neg':
      val-=step
  xs= [i for i in range(len(ys))]
  return np.array(xs, dtype= np.float64) , np.array(ys, dtype= np.float64)
```

```
xs, ys = create_dataset(40,10,2,'pos') #change the variance (to different values and false) and see the effect on graph and rsquare value
m = bestFitSlope(xs, ys)
b = intercept(xs,ys,m)

predict_x = 8
predict_y = m*predict_x + b

regressionLine = [(m*x)+b for x in xs]
r_squared = coeff(ys, regressionLine)
print(r_squared)


plt.scatter(predict_x, predict_y, s=100, color = 'r')
plt.scatter(xs,ys)
plt.plot(xs,regressionLine)
```
